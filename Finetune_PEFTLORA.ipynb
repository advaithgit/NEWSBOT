```python
from datasets import load_dataset
from transformers import pipeline

from transformers import AutoModelForSeq2SeqLM, AutoTokenizer
import torch
```

```python
device = 'gpu'
model_ckpt = 'facebook/bart-large-cnn'
tokenizer = AutoTokenizer.from_pretrained(model_ckpt)
model = AutoModelForSeq2SeqLM.from_pretrained(model_ckpt)
```

```python
samsum = load_dataset('Samsung/samsum')
samsum
```

```python
print(samsum['train'][4]['dialogue'])
```

```python
def get_feature(batch):
  encodings = tokenizer(batch['dialogue'], text_target=batch['summary'],
                        max_length=1024, truncation=True)

  encodings = {'input_ids': encodings['input_ids'],
               'attention_mask': encodings['attention_mask'],
               'labels': encodings['labels']}

  return encodings
```

```python
samsum_pt = samsum.map(get_feature, batched=True)
```

```python
columns = ['input_ids', 'labels', 'attention_mask']
samsum_pt.set_format(type='torch', columns=columns)
```

```python
from transformers import DataCollatorForSeq2Seq
data_collator = DataCollatorForSeq2Seq(tokenizer, model=model)
```

```python
from huggingface_hub import notebook_login

notebook_login()
```

```python
from transformers import TrainingArguments, Trainer

training_args = TrainingArguments(
    output_dir = 'bart_samsum',
    hub_model_id='Advaith1612/bart_convo-summarizer',
    num_train_epochs=1,
    warmup_steps = 500,
    per_device_train_batch_size=4,
    per_device_eval_batch_size=4,
    weight_decay = 0.01,
    logging_steps = 10,
    evaluation_strategy = 'steps',
    eval_steps=500,
    save_steps=1e6,
    gradient_accumulation_steps=16
)

trainer = Trainer(model=model, args=training_args, tokenizer=tokenizer, data_collator=data_collator,
                  train_dataset = samsum_pt['train'], eval_dataset = samsum_pt['validation'])
```

```python
trainer.train()
```

```python
trainer.save_model('bart_samsum_model')
```

```python
from datasets import load_dataset
samsum = load_dataset('samsum')
samsum
```

```python
from peft import LoraConfig,get_peft_model,TaskType

lora_config=LoraConfig(
    r=32,
    lora_alpha=32,
    lora_dropout=0.05,
    bias="none",
    task_type=TaskType.SEQ_2_SEQ_LM
)
```

```python
# Load model directly
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("Advaith1612/bart_convo-summarizer")
model = AutoModelForSeq2SeqLM.from_pretrained("Advaith1612/bart_convo-summarizer")
```

```python
peft_model=get_peft_model(model,lora_config)
```

```python
from huggingface_hub import notebook_login

notebook_login()
```

```python
print(samsum)
```

```python
def tokenize_input(example):
    start_prompt = "Summarise the following conversation: \n\n"
    end_prompt = "\n\nSummary:"
    prompt = [start_prompt + dialogue + end_prompt for dialogue in example['dialogue']]

    inputs = tokenizer(prompt, padding='max_length', truncation=True, max_length=1024, return_tensors='pt')
    labels = tokenizer(example['summary'], padding='max_length', truncation=True, max_length=1024, return_tensors='pt')

    example['input_ids'] = inputs['input_ids']
    example['attention_mask'] = inputs['attention_mask']
    example['labels'] = labels['input_ids']

    return example

# Ensure the padding token is defined
tokenizer.pad_token = tokenizer.eos_token

# Tokenize the dataset
tokenized_datasets = samsum.map(tokenize_input, batched=True)

# Remove unwanted columns
tokenized_datasets = tokenized_datasets.remove_columns(['id', 'dialogue', 'summary'])

# Filter the dataset
tokenized_datasets = tokenized_datasets.filter(lambda example, index: index % 100 == 0, with_indices=True)

print(tokenized_datasets)
```

```python
from transformers import TrainingArguments, Trainer

peft_training_args = TrainingArguments(
    output_dir = 'bart_samsum_peft',
    hub_model_id='Advaith1612/bart-convo-summarizer-peft',
    learning_rate=1e-5,
    num_train_epochs=5,
    weight_decay=0.01,
    auto_find_batch_size=True,
    logging_steps = 10,
    evaluation_strategy = 'epoch'
    )

peft_trainer = Trainer(model=peft_model,
                       args=peft_training_args,
                       train_dataset=tokenized_datasets['train'],
                       eval_dataset=tokenized_datasets['validation']
                       )
```

```python
peft_model.print_trainable_parameters()
```

```python
peft_trainer.train()
```

```python
peft_trainer.push_to_hub("Advaith1612/bart-convo-summarizer-peft")
```

#PEFT TESTING

```python
from peft import PeftModel, PeftConfig
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM


tokenizer = AutoTokenizer.from_pretrained("Advaith1612/bart_convo-summarizer")
peft_model_base = AutoModelForSeq2SeqLM.from_pretrained("Advaith1612/bart_convo-summarizer")

loaded_peft_model=PeftModel.from_pretrained(peft_model_base,"Advaith1612/bart-convo-summarizer-peft",is_trainable=False)
```

```python
def generate_summary(input,llm):
  input_prompt=f"""Summarise the following conversation: \n\n{input}\n\nSummary:"""
  input_ids=tokenizer(input,return_tensors='pt')
  tokenized_output=llm.generate(input_ids=input_ids['input_ids'],min_length=30,max_length=200)
  output=tokenizer.decode(tokenized_output[0],skip_special_tokens=True)
  return output

```

```python
sample=samsum['validation'][0]['dialogue']
label=samsum['validation'][0]['summary']

output=generate_summary(sample,loaded_peft_model)

```
