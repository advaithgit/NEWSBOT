```python
# Download NLTK data
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')

# Initialize NLTK components
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

# Initialize transformers pipelines
sentiment_analyzer = pipeline("sentiment-analysis")
text_classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

# Initialize Spacy for NER
nlp = spacy.load("en_core_web_sm")

# Function to perform Named Entity Recognition (NER)
def perform_ner(text):
    doc = nlp(text)
    entities = [(ent.text, ent.label_) for ent in doc.ents]
    return entities

# Function to perform Sentiment Analysis
def perform_sentiment_analysis(text):
    result = sentiment_analyzer(text)[0]
    sentiment = result['label']
    score = result['score']
    return f"{sentiment} ({score:.2f})"

# Function to extract Keywords
def extract_keywords(text):
    words = word_tokenize(text)
    words = [lemmatizer.lemmatize(word.lower()) for word in words if word.isalnum() and word.lower() not in stop_words]
    keywords = nltk.FreqDist(words).most_common(10)
    return [keyword for keyword, _ in keywords]

# Function to perform Text Classification
def classify_text(text):
    candidate_labels = ["business", "politics", "technology", "health", "entertainment", "sports", "science"]
    result = text_classifier(text, candidate_labels)
    return result['labels'][0]

# Load Excel file

# Lists to store results
ners = []
sentiments = []
keywords = []
text_classifications = []

# Iterate over each row in the dataframe
for index, row in df.iterrows():
    snippet = row['Summary']

    # Perform NER
    ner_result = perform_ner(snippet)
    ners.append(ner_result)

    # Perform Sentiment Analysis
    sentiment_result = perform_sentiment_analysis(snippet)
    sentiments.append(sentiment_result)

    # Extract Keywords
    keywords_result = extract_keywords(snippet)
    keywords.append(keywords_result)

    # Perform Text Classification
    text_classification_result = classify_text(snippet)
    text_classifications.append(text_classification_result)

# Add the results as new columns in the dataframe
df['NER'] = ners
df['Sentiment'] = sentiments
df['Keywords'] = keywords
df['Text Classification'] = text_classifications

# Save the updated dataframe to a new Excel file
df.to_excel('processed_file_with_ner_sentiment_keywords_classification.xlsx', index=False)

# Print the dataframe with the new columns
print(df)
```

```python
import pandas as pd
from langchain_chroma import Chroma
from langchain_community.embeddings.sentence_transformer import SentenceTransformerEmbeddings

# Define the input Excel file path
data_file = 'processed_file_with_ner_sentiment_keywords_classification.xlsx'

# Read the Excel file into a DataFrame
data_df = pd.read_excel(data_file)

# Initialize the HuggingFace embeddings using the new import path
hf_embed = SentenceTransformerEmbeddings(model_name="all-MiniLM-L6-v2")

# Function to format the content based on the given row data
def format_content(row):
    content = row['content']
    ner = eval(row['NER'])
    sentiment = row['Sentiment'].split()[0]  # Extract the sentiment without the score
    keywords = ', '.join(eval(row['Keywords']))  # Convert string representation of list to list
    text_classification = row['Text Classification']

    ner_formatted = "\n".join([f"{entity[1]}: {entity[0]}" for entity in ner])

    formatted_text = f"""Content: {content}. NER: {ner_formatted}. Sentiment: {sentiment}. Keywords: {keywords}. Classification: {text_classification}"""
    return formatted_text.strip()

# Create a list to hold the formatted texts
formatted_texts = []

# Format the content for each row
for index, row in data_df.iterrows():
    formatted_text = format_content(row)
    formatted_texts.append(formatted_text)

# Create Chroma vector store from the formatted texts
db = Chroma.from_texts(formatted_texts, hf_embed, persist_directory="kai2")
```

```python
query = "Cryptocurrency markets"
docs = db.similarity_search(query,k=3)
#string=str(docs[0])
#print(string[22:-1])
print((docs)[2])
```

```python
chat = ""
while True:
    choice = int(input("ENTER CHOICE: 1. Query, 2. Question Answering, 3. Exit: "))
    if choice == 3:
        break


    if choice == 1:
        query = input("Enter topic on what you want to query: ")
        docs = db.similarity_search(query)

        if docs:
            string = str(docs[0])[22:-1]
            cleaned_text = string.replace('"', '').replace('\n', ' ')
            context = cleaned_text
            print(context)
            prev = f"User Queried This: {query}. Data Base Retrieved This: {context}"
            chat += "\n" + prev
            #print(chat)
        else:
            print("No relevant documents found.")

    elif choice == 2:
        question = input("Enter question: ")
        docs = db.similarity_search(question)

        if docs:
            string = str(docs[0])[22:-1]
            cleaned_text = string.replace('"', '').replace('\n', ' ')
            context = cleaned_text
            #print(chat)
            prevchat = convosummarizer(chat)[0]['generated_text']
            prompt = f"""You are a chatbot that remembers the previous chat with the user. Answer based on the given context and previous conversation alone. Do not rely on your pre-trained knowledge. If you do not know the answer, say so instead of making one up. If the context is not relevant to question, say you didn't find any relevant document, don't make up an answer. Question: {question}. Context: {context}. Previous Chat: {prevchat}"""
            #print(prompt)
            messages = [
                {"role": "user", "content": prompt},
            ]
            response = pipe(messages)
            answer = next(item['content'] for item in response[0]['generated_text'] if item['role'] == 'assistant')
            print(answer)
            prev = f"User: {question} Chatbot: {answer}"
            chat += "\n" + prev
        else:
            print("No relevant documents found.")

    else:
        print("Invalid choice. Please try again.")
```
